```python
provider = IBMProvider()
print(provider.backends())
```

```python
backend = provider.get_backend('ibm_brisbane')

backend_config = backend.configuration()
backend_defaults = backend.defaults()
backend_properties = backend.properties()
dt = backend_config.dt

qubit = 109
cbit = 0
num_qubits = int(qubit+1)
num_cbits = 1
weight = 1
```

```python
qubit_01_freq = backend_defaults.qubit_freq_est[qubit]
qubit_anharmonicty = backend_properties.qubits[qubit][3].value*GHz
qubit_12_freq = qubit_01_freq + qubit_anharmonicty

p12_amp = 0.22289722723304814
```

```python
with pulse.build(backend=backend) as hp12_sched:
    drive_chan = pulse.drive_channel(qubit)
    pulse.set_frequency(qubit_12_freq, drive_chan)
    pulse.play(pulse.Drag(duration=96, amp=p12_amp/2, sigma=24, beta=-0.4, name='hp12'), drive_chan)
    
hp12_gate = Gate(r'X_{\pi/2}^{(12)}', weight, [])
    
with pulse.build(backend=backend) as p12_sched:
    drive_chan = qiskit.pulse.drive_channel(qubit)
    pulse.set_frequency(qubit_12_freq, drive_chan)
    pulse.play(pulse.Drag(duration=96, amp=p12_amp, sigma=24, beta=-0.4, name='p12'), drive_chan)

p12_gate = Gate(r'X_{\pi}^{(12)}', weight, [])
```

```python
with pulse.build(backend=backend) as delay_10us:
    drive_chan = pulse.drive_channel(qubit)
    pulse.delay(20000, drive_chan)
    
with pulse.build(backend=backend) as delay_5us:
    drive_chan = pulse.drive_channel(qubit)
    pulse.delay(10000, drive_chan)
    
delay_gate = Gate('Delay', weight, [])
```

## Discriminator

```python
ground_circ = QuantumCircuit(num_qubits, num_cbits)
ground_circ.append(delay_gate, [qubit])
ground_circ.measure(qubit, cbit)
ground_circ.add_calibration(delay_gate, [qubit], delay_10us)

first_excited_state_circ = QuantumCircuit(num_qubits, num_cbits)
first_excited_state_circ.append(delay_gate, [qubit])
first_excited_state_circ.x(qubit)
first_excited_state_circ.measure(qubit, cbit)
first_excited_state_circ.add_calibration(delay_gate, [qubit], delay_10us)

second_excited_state_circ = QuantumCircuit(num_qubits, num_cbits)
second_excited_state_circ.append(delay_gate, [qubit])
second_excited_state_circ.x(qubit)
second_excited_state_circ.append(p12_gate, [qubit])
second_excited_state_circ.measure(qubit, cbit)
second_excited_state_circ.add_calibration(p12_gate, [qubit], p12_sched)
second_excited_state_circ.add_calibration(delay_gate, [qubit], delay_10us)

discrim_circs = qiskit.transpile([ground_circ, first_excited_state_circ, second_excited_state_circ], backend=backend)
```

## The Amplified Phase Error Sequence

[1]: Erik Lucero *et al*. Reduced phase error through optimized control of a superconducting qubit. *Phys. Rev. A* **82**, 042339 (2010). [https://link.aps.org/doi/10.1103/PhysRevA.82.042339](https://link.aps.org/doi/10.1103/PhysRevA.82.042339).

The phase error expressed in circuit language is 

\begin{align}
X'_{\pi/2} = e^{i\epsilon'} Z_{\epsilon} X_{\pi/2} Z_{\epsilon} 
\end{align}

This phase error $\epsilon$ is a function of both angle rotation $\theta$ and gate time $t_g$ (See also McKay's Virtual Z gates). This means if we can calibrate $X_{\pi/2}$ gate and construct every Clifford rotation from them, we can guarantee that the Clifford phase errors are suppressed, since the Hadamard gate and $S$ gate generate the Clifford group. 

**Alternatively**, how about using the Virtual $Z$ gate to correct for the phase error, and then use both $Y$ and $Z$ for leakage? Nevermind, we'll explore this idea later.

Now we seek a method to amplify the phase error $\epsilon$. Consider a concatenation of four $\pi/2$ pulses that amounts to a pseudo-$I$ operator. However, the phase errors in this case cancel each other and result in a global phase (**Note**: global phase of the qutrit or the qubit?). Consider another sequence of concatenated positive and negative $\pi/2$ rotations, this results in 

\begin{align}
I'_{\pi/2}= \begin{pmatrix}
1-i\epsilon & \epsilon \\
-\epsilon & 1-3i\epsilon 
\end{pmatrix}
\end{align}

An $n$ successive application of $I'_{\pi/2}$ results in an approximated $Z$ gate

\begin{align}
I'_{\pi/2} \approx Z_{2n\epsilon}
\end{align}

## DRAG YZ probe + APE

Amplified phase error (APE) pulse sequence in combination with YZ components of DRAG.

We'll have a bit of detuning, from $-4.5$ MHz to $1$ MHz.

```python
detuning_MHz = np.linspace(-4.5, 1.9, 15)
detuning_MHz = detuning_MHz*MHz
```

Previous experiments show that good $\beta$ are typically negative. We'll sweep from $-5.0$ to $3.0$ (A.U.).

```python
scaled_quadrature = np.linspace(-5.0, +3.0, 30)
```

```python
YZ = np.zeros([len(detuning_MHz), len(scaled_quadrature)], dtype=tuple)

for i in range(len(detuning_MHz)):
    for j in range(len(scaled_quadrature)):
        YZ[i, j] = (detuning_MHz[i], scaled_quadrature[j])
```

```python
YZ.shape
```

```python
YZ_circs = []

for i in range(len(detuning_MHz)):
    for j in range(len(scaled_quadrature)):
        
        pulse_detuning = YZ[i, j][0]
        beta = YZ[i, j][1]
        
        YZ_circ = QuantumCircuit(num_qubits, num_cbits)
        YZ_circ.append(delay_gate, [qubit])
        YZ_circ.x(qubit) # prepare |1> by default X gate of IBM's
        YZ_circ.append(hp12_gate, [qubit]) # here we insert a pre-calibrated $\pi$ pulse. This should be re-examined. 
        
        with pulse.build(backend=backend) as pseudo_identity_sched:
            drive_chan = pulse.drive_channel(qubit)
            pulse.set_frequency(qubit_12_freq+pulse_detuning, drive_chan)
            pulse.play(pulse.Drag(duration=96, amp=+1*p12_amp/2, sigma=24, beta=beta, name='PLUS_hp12_YZ'), drive_chan)
            pulse.play(pulse.Drag(duration=96, amp=-1*p12_amp/2, sigma=24, beta=beta, name='MINUS_hp12_YZ'), drive_chan)

        pseudo_identity_gate = Gate(r'Xpi/2-Xmpi/2', weight, [])
        
        # n = 3 i.e. 1 pseudo sequence of 3*I
        YZ_circ.append(pseudo_identity_gate, [qubit])
        YZ_circ.append(pseudo_identity_gate, [qubit])
        YZ_circ.append(pseudo_identity_gate, [qubit])
        
        YZ_circ.append(hp12_gate, [qubit])
        
        YZ_circ.measure(qubit, cbit)
        
        YZ_circ.add_calibration(delay_gate, [qubit], delay_sched)
        YZ_circ.add_calibration(hp12_gate, [qubit], hp12_sched)
        YZ_circ.add_calibration(pseudo_identity_gate, [qubit], pseudo_identity_sched)
        
        YZ_circs.append(YZ_circ)
```

```python
package = []

package = discrim_circs + YZ_circs
```

```python
YZ_exp = backend.run(package, meas_level=1, meas_return='single', shots=1024)
```
